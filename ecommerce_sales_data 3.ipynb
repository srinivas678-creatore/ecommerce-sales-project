```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df
```

```python
df = df.sort_values('Order Date')
df
```

```python
daily_sales = df.groupby('Order Date')['Sales'].sum()
```

```python

plt.figure(figsize=(10,5))
plt.plot(daily_sales.index, daily_sales.values)

plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.title("Daily Sales Trend")
```

```python
region_sales = df.groupby('Region')['Sales'].sum()

plt.figure(figsize=(8,5))
plt.bar(region_sales.index, region_sales.values)

plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.title("Region-wise Sales Comparison")
```

```python
df.groupby('Category')['Profit'].sum()
```

```python

category_profit = df.groupby('Category')['Profit'].sum()

plt.figure(figsize=(7,7))
plt.pie(category_profit.values,
        labels=category_profit.index,
        autopct='%1.1f%%')

plt.title("Category-wise Profit Distribution")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64','float64'])

sns.heatmap(numeric_df.corr(), annot=True)
plt.title("Heat Map ")
```

```python

plt.figure(figsize=(8,5))
sns.boxplot(x=df['Quantity'],y=df["Sales"])

plt.title("Sales Distribution - Box Plot")
plt.ylabel("Sales Amount")
plt.show()
```
